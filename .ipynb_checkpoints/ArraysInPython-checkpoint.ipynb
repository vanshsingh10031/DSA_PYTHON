# Python Array Basic For Basic Issue 
Arrays in Python are used to store multiple values of the same data type in a single variable. They allow efficient storage and easy access to elements using index positions, making them useful for handling numeric data.

### Load Necessary Libraries

```python
import array as arr

# initial size
n = 1

a = arr.array('i', [0] * n)

for i in range(n):
    a[i] = int(input("Enter your number: "))

while True:
    choice = input("Do you want to add more numbers? (y/n): ").lower()

    if choice == 'y':
        more = int(input("How many more numbers do you want to add? "))

        for _ in range(more):
            a.append(int(input("Enter your number: ")))

    elif choice == 'n':
        break

    else:
        print("Invalid choice. Please enter y or n.")

print("Final array elements:")
print(a)
```

```python
import array as arr


def insertatpos(a, pos, val, maxsize):
    if len(a) >= maxsize:
        print("Array is full. Cannot insert.")
        return

    if pos < 1 or pos > len(a) + 1:
        print("Invalid position.")
        return

    a.append(0)

    for i in range(len(a) - 1, pos - 1, -1):
        a[i] = a[i - 1]

    a[pos - 1] = val
    print("Array after insertion:", a)


n = int(input("Enter the maximum number of elements allowed: "))
actual_n = int(input("Enter how many elements you want to add initially: "))

a = arr.array('i')

if actual_n > n:
    print("Error: Initial count exceeds maximum size.")
else:
    for i in range(actual_n):
        a.append(int(input(f"Enter element {i + 1}: ")))

    print("Current array:", a)

    pos = int(input("Enter the position where you want to insert (1-based index): "))
    val = int(input("Enter the value to insert: "))

    insertatpos(a, pos, val, n)
```

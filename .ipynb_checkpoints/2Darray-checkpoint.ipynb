```python
#Wap to Make a 2D Array
import array as arr

arr = [
    [1, 2],
    [3, 4]
]
print(arr);

# Traversing through an 2D Array
for i in range(len(arr)):          
    for j in range(len(arr[i])):   
        print(arr[i][j], end=" ")
    print()

#addition of 2 , 2d arrays 
arr1 = [
    [1, 2],
    [3, 4]
]

arr2 = [
    [1, 2],
    [3, 4]
]

#Adding of arr1 and arr2 
# Traversing through an 2D Array
for i in range(len(arr1)):          
    for j in range(len(arr2[i])):   
        print(arr2[i][j], end=" ")
    print()
```

```python
#addition of 2 , 2d arrays 
arr1 = [
    [1, 2],
    [3, 4]
]

arr2 = [
    [5, 6],
    [7, 8]
]


#Adding of arr1 and arr2
for i in range(len(arr1)):          
    for j in range(len(arr2[i])):
        print(i)
        print(j)
        print(arr1[i])
        print(arr2[j])
        print(arr1[i][j] + arr2[i][j])
       # print(arr2[i][j], end=" ")
    print()



#Substraction of arr1 and arr2
for i in range(len(arr1)):          
    for j in range(len(arr2[i])):
        print(i)
        print(j)
        print(arr1[i])
        print(arr2[j])
        print(arr1[i][j] - arr2[i][j])
       # print(arr2[i][j], end=" ")
    print()
```

```python
# WAP to create an Identity Matrix

n = 2   # size of matrix

identity = []

for i in range(n):
    row = []
    for j in range(n):
        if i == j:
            row.append(1)
        else:
            row.append(0)
    identity.append(row)

print(identity)
```

```python
# WAP to create transpose of a matrix

matrix = [
    [5, 6, 9],
    [7, 8, 10]
]
transpose = []

for i in range(len(matrix[0])):      # columns of original
    row = []
    for j in range(len(matrix)):     # rows of original
        row.append(matrix[j][i])
    transpose.append(row)

for row in transpose:
    print(row)
```

```python
#initilization of array
array = [6, 5, 2, 9, 2, 5, 0]

#taking in length 
n = len(array)
# print(n)

for i in range(n - 1):  
    #print(i)
    for j in range(n - i - 1):  
        #print(j)
        if array[j] > array[j + 1]:
            print(array[j] ,array[j+1] )
            array[j], array[j + 1] = array[j + 1], array[j]

print(array)
```

```python
# Deletion of an Array
deletionarray = [10, 20, 30, 40]
deletionposition = 3  # 1-based index

for i in range(deletionposition - 1, len(deletionarray) - 1):
    deletionarray[i] = deletionarray[i + 1]
    
print(deletionarray)
```

```python
# Finding the Duplicate in an Array
array = [1, 2, 3, 4, 5, 6]

for i in range(len(array)):
    print(array[i])
```

```python
#Finding Duplicate Elements in a Given Array
array = [1, 2, 3, 4, 2, 5, 6]
print("Length :",len(array))
for i in range(len(array)):
    #print(i)
    for j in range(i + 1, len(array)):
        if array[i] == array[j]:
            print("Duplicate:", array[i])
```

```python
# Making an Array with Duplicate Values
array = [1, 2, 3, 4,3,4, 2, 5, 6]
a = []

print("Length:", len(array))

for i in range(len(array)):
    for j in range(i + 1, len(array)):
        if array[i] == array[j]:
            a.append(array[i])
            print("Duplicate:", array[i])

print("Duplicated Array:", a)
```

```python
# Linear Searching in an Array
array = [1, 2, 3, 4, 5, 6]
value = 12

for i in range(len(array)):
    if array[i] == value:
        print("Target Found")
    else :
        print("Target not Found")
        break
```

```python
# Reversing the Array
reversingarray = [1, 2, 3, 4, 5]
array = []

for i in range(len(reversingarray) - 1, -1, -1):
    print(reversingarray[i])
    array.append(reversingarray[i])

print(array)

if reversingarray == array:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
#Creating a class to create a node
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
# Linked List Class
N1 = Node(10)
N2 = Node(20)

# Linking Nodes
N1.next = N2
```

```python
#code for Displaying the Linked List
def display(head):
    current = head
    while current is not None:
        print(current.data)
        current = current.next

display(N1)
```

```python
#code for Traversing the Linked List
def transverse(head):
    current = head
    while current is not None:
        print(current.data)
        current = current.next

transverse(N1)  
```

```python
#Inserting at the beginning of the Linked List
def insertatbeg(Head , data):
    newnode = Node(data)
    newnode.next = Head
    head = newnode
    return head

N1 = insertatbeg(N1, 5)
display(N1)
```

```python
#inserting at the End
def insertingatend(head , data):
    newnode = Node(data)
    if head is None:
        head = newnode
        return head
    current = head
    while current.next is not None:
        current = current.next 
    current.next = newnode
    return head
N1 = insertingatend(N1, 25)
display(N1)
```

```python
#inserting at the position
def insertingatpostion(head, data, position):
    newnode = Node(data)
    if head is None:
        head = newnode
        return head
    current = head
    #firstly to count the number of nodes in the linked list
    count = 0
    while current.next is not None:
        count += 1
        current = current.next 
    #then to check if the position is valid or not
    if position < 0 or position > count:
        print("Invalid Position")
        return head 
    
    current = head
    for i in range(position-1):
        current = current.next
    newnode.next = current.next
    current.next = newnode
        
    return head
N1 = insertingatpostion(N1, 25, 2)
N1 = insertingatpostion(N1, 100, 3)
display(N1)
```

```python
# Insert at Middle (Optimized: Fast-Slow Pointer)
def insertingatmiddle(head, data):
    newnode = Node(data)

    # Case: Empty list
    if head is None:
        return newnode

    slow = head
    fast = head

    # Move fast 2 steps, slow 1 step
    while fast.next and fast.next.next:
        slow = slow.next
        fast = fast.next.next

    # Insert after middle
    newnode.next = slow.next
    slow.next = newnode

    return head



# Creating initial nodes
N1 = Node(10)
N2 = Node(20)
N3 = Node(30)
N4 = Node(40)

# Linking nodes
N1.next = N2
N2.next = N3
N3.next = N4

print("Original List:")
display(N1)


# Insert at Beginning
N1 = insertatbeg(N1, 5)
print("After inserting at beginning (5):")
display(N1)


# Insert at Middle
N1 = insertingatmiddle(N1, 50)
print("After inserting at middle (50):")
display(N1)
```

```python
#deleting an element from the end
def deleteatend(head):
    #validating if the list is empty or not
    if head is None:
        print("List is empty")
        return None
    #validating if the list has only one node
    if head.next is None:
        return None
    
    #creating a pointer to traverse the list
    current = head

    #running the loop until the second last node is reached
    while current.next.next is not None:
        current = current.next

    #deleting the last node by making the next of second last node as None
    current.next = None
    return head
```

```python
def delete_middle(head):
    if head is None or head.next is None:
        return None

    # count nodes
    count = 0
    temp = head
    while temp:
        count += 1
        temp = temp.next

    middle = count // 2 + 1

    temp = head
    for _ in range(middle - 2):
        temp = temp.next

    temp.next = temp.next.next

    return head


```

```python
#Searching an Element in the linked list
def search(head , key):
    #validating if the list is empty or not
    if head is None:
        print("List is empty")
        return None
    current = head
    while current is not None:
        if current.data == key:
            print("Element Found")
            return current
        current = current.next
    print("Element Not Found")
    return None

search(N1,20);
display(N1);
```

```python
#code for counting the number of nodes in the linked list
def countnodes(head):
    count = 0
    current = head
    while current is not None:
        count += 1
        current = current.next
    return count

countnodes=countnodes(N1)
print("Number of nodes in the linked list:", countnodes)
```

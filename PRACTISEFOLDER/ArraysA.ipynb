```python
Size = int(input("Enter the Size of the Array = "))

#inserting at a postion 
d = arr.array('i', [0]*Size)

for i in range(len(d)):
    d[i] = int(input("Enter A number = "))


#Making A function to do that(Passing position and Value)
def insertatpostion(pos , val):
    
    for i in range(len(d)):
        if pos<0 or pos>=len(d):
            print("Position is InValid")
             
        if i == pos:
            d[i] = val
            break
   
    
position = int(input("Enter the Position to insert = "))
Value = int(input("Enter the Value to insert = "))
insertatpostion(position-1, Value)
print(d)
```

```python
#Printing the Array in Reverse Order
Size = int(input("Enter the Size of the Array = "))

#inserting at a postion 
e= arr.array('i', [0]*Size)

for i in range(len(e)):
    e[i] = int(input("Enter A number = "))


#Making a loop to print the array in reverse order
for i in range(len(e)-1, -1, -1):
    print(e[i]) 
```
